4.Escribe una función suma_y_promedio(a, b, c) que reciba tres números, calcule su suma y su promedio, y regrese ambos valores.


```python
def suma_y_promedio(a, b, c):
    suma = a + b + c
    promedio = suma / 3
    return suma, promedio

# Ejemplo
print(suma_y_promedio(2, 4, 6))
```

5.Escribe una función es_par(numero) que devuelva "Par" si el número es par, o "Impar" si es impar.

```python
def es_par(numero):
    if numero % 2 == 0:
        return "Par"
    else:
        return "Impar"

# Ejemplo
print(es_par(7))
```

6.Crea una función clasificar_edad(edad) que devuelva:



*   "Niño" si la edad es menor a 12,

*   "Adolescente" si está entre 12 y 17,

*   "Adulto" si es 18 o más.

```python
def clasificar_edad(edad):
    if edad < 12:
        return "Niño"
    elif edad < 18:
        return "Adolescente"
    else:
        return "Adulto"

# Ejemplo
print(clasificar_edad(15))  # Resultado: Adolescente
```

7. Escribe una función mayor_de_tres(a, b, c) que reciba tres números y devuelva el mayor de ellos.

```python
def mayor_de_tres(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= c:
        return b
    else:
        return c
# Ejemplo
print(mayor_de_tres(4, 10, 6))
```

8.Escribe una función esta_en_rango(x) que devuelva True si x está entre 0 y 100 (incluyendo ambos), y False si no lo está.

```python
def esta_en_rango(x):
    return 0 <= x <= 100

# Ejemplo
print(esta_en_rango(50))
print(esta_en_rango(150))
```

9.Escribe una función resumen_lista(lista) que imprima:

El número máximo de la lista

El número mínimo

El número de elementos

La suma total

(Usa las funciones max(), min(), len() y sum())

```python
def resumen_lista(lista):
    print("Máximo:", max(lista))
    print("Mínimo:", min(lista))
    print("Cantidad de elementos:", len(lista))
    print("Suma total:", sum(lista))

# Ejemplo
resumen_lista([5, 7, 3, 9, 2])
```

10.Crea una función es_mayor_y_par(numero) que devuelva True si el número es mayor a 10 y además es par. En cualquier otro caso, que devuelva False.

11.Escribe una función contar_negativos(lista) que use NumPy para contar cuántos números negativos hay en una lista de números.

```python
import numpy as np

def contar_negativos(lista):
    arr = np.array(lista)
    return np.sum(arr < 0)

# Ejemplo
print(contar_negativos([3, -1, -7, 4, 0,-9,12,-34]))
```

12.Crea una función estadisticas(lista) que reciba una lista de números y devuelva una tupla con la media y la mediana, usando NumPy.

```python
import numpy as np

def estadisticas(lista):
    arr = np.array(lista)
    return np.mean(arr), np.median(arr)

# Ejemplo
print(estadisticas([12, 25, 3, 4, 5, 12,32,]))
```

13.Escribe una función clasificar_calificacion(puntuacion) que reciba una calificación del 0 al 100 y devuelva:


*   "Excelente" si es mayor o igual a 90
*   "Bien" si está entre 70 y 89
*   "Regular" si está entre 50 y 69
*   "Insuficiente" si es menor a 50


    

   

    

    

```python
def clasificar_calificacion(puntuacion):
    if puntuacion >= 90:
        return "Excelente"
    elif puntuacion >= 70:
        return "Bien"
    elif puntuacion >= 50:
        return "Regular"
    else:
        return "Insuficiente"

# Ejemplo
print(clasificar_calificacion(85))
```
