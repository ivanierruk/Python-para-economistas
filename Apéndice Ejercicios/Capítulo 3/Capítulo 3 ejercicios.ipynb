 1.Usa un bucle for y la función range() para imprimir los números del 1 al 10.

3.Usa un bucle for para imprimir cada letra de la palabra "Python".

4.Crea un bucle while que cuente del 1 al 8.



```python
i = 1

while i <= 8:
    print(i)
    i += 1
```

5. Cuenta cuántas veces aparece el número 3 en la lista [1, 3, 5, 3, 7, 3, 9].


```python
lista = [1, 3, 5, 3, 7, 3, 9]
contador = 0

for numero in lista:
    if numero == 3:
        contador += 1

print("El número 3 aparece", contador, "veces.")
```

6.Imprime una tabla de conversión de centímetros a pulgadas del 10 al 100, de 10 en 10 (ten en cuenta que 1 pulgada = 2.54 cm).



```python
for cm in range(10, 101, 10):
    pulgadas = cm / 2.54

    print(f"{cm} cm = {pulgadas:.2f} pulgadas")
```

7.Usa un bucle for para sumar los dígitos del número 12345.


```python
numero = "12345"
suma = 0

for digito in numero:
    suma += int(digito)

print("La suma de los dígitos es:", suma)
```

8.Usa bucles anidados para imprimir las tablas de multiplicar del 1 al 10



```python
for i in range(1, 11):
    print(f"Tabla del {i}:")

    for j in range(1, 11):
        print(f"{i} x {j} = {i * j}")
    print()

```

10.Cuenta cuántas vocales tiene la palabra "Economía".

```python
palabra = "Economía"
vocales = "aeiouAEIOU"
contador = 0

for letra in palabra:
    if letra in vocales:
        contador += 1

print("La palabra tiene", contador, "vocales.")
```

11.Dada una lista de palabras, crea una nueva lista con sus longitudes de sus letras.

```python
#Ejemplo: palabras[hola, mundo, Python]
palabras = ["hola", "mundo", "Python"]
longitudes = []

for palabra in palabras:
    longitudes.append(len(palabra))

print(longitudes)
```

13.Dadas dos listas del mismo tamaño, crea una nueva lista con el producto de sus elementos correspondientes.


*   a= [1, 2, 3]
*   b= [4, 5, 6]



```python
a = [1, 2, 3]
b = [4, 5, 6]

producto = [a[i] * b[i] for i in range(len(a))]

print(producto)
```
