### **Ejercicio 1: Simulación y Visualización de Lanzamientos de un Dado**

**Instrucciones:**

1.  **Importa las librerías** `numpy` como `np` y `matplotlib.pyplot` como `plt`.

2.  **Define una función** llamada `analizar_lanzamientos_dado` que acepte un argumento: `numero_de_lanzamientos`.

3.  **Simulación Vectorizada**: Dentro de la función, utiliza `np.random.randint()` para generar un array que simule todos los `numero_de_lanzamientos` de un dado de 6 caras (valores del 1 al 6).

4.  **Análisis de Frecuencias**:
    * Imprime un resumen que indique cuántas veces salió cada número (del 1 al 6). Puedes usar un bucle `for` y una máscara booleana de NumPy (`np.sum(resultados == cara)`) para hacer este conteo.

5.  **Visualización con un Histograma**:
    * Utiliza `plt.hist()` para crear un histograma de los resultados de la simulación.
    * **Configura los `bins` (contenedores)** para que cada barra represente una cara del dado. Una buena técnica es: `bins = np.arange(1, 8) - 0.5`. Esto centra las barras en los números 1, 2, 3, etc.
    * **Personaliza el gráfico**: Añade un título (ej. "Distribución de Lanzamientos de un Dado"), y etiquetas para los ejes X ("Resultado") e Y ("Frecuencia").

6.  **Muestra el gráfico** usando `plt.show()`.

7.  **Prueba la función** llamándola para simular al menos `10,000` lanzamientos, para que la distribución sea claramente visible.

```python
# 1. Importar las librerías
import numpy as np
import matplotlib.pyplot as plt

# 2. Definir la función de análisis
def analizar_lanzamientos_dado(numero_de_lanzamientos):
    """
    Simula lanzamientos de un dado, analiza las frecuencias
    y visualiza los resultados con un histograma.
    """
    print(f"--- Simulación de {numero_de_lanzamientos:,} lanzamientos de un dado ---")

    # 3. Simulación Vectorizada
    # Genera enteros aleatorios entre 1 y 7 (sin incluir el 7)
    resultados = np.random.randint(1, 7, size=numero_de_lanzamientos)

    # 4. Análisis de Frecuencias
    print("\nFrecuencia de cada resultado:")
    for cara in range(1, 7):
        conteo = np.sum(resultados == cara)
        porcentaje = (conteo / numero_de_lanzamientos) * 100
        print(f"  Cara {cara}: {conteo} veces ({porcentaje:.2f}%)")

    # 5. Visualización con un Histograma
    # Bins de 0.5 a 1.5, 1.5 a 2.5, etc., para centrar las barras en los enteros
    bins = np.arange(1, 8) - 0.5

    plt.hist(resultados, bins=bins, color='skyblue', edgecolor='black', rwidth=0.8)

    # Personalización del gráfico
    plt.title(f'Distribución de {numero_de_lanzamientos:,} Lanzamientos de Dado')
    plt.xlabel('Resultado del Dado')
    plt.ylabel('Frecuencia Absoluta')
    plt.xticks(range(1, 7)) # Asegura que las marcas del eje X sean 1, 2, 3, 4, 5, 6
    plt.grid(axis='y', alpha=0.75)

    # 6. Mostrar el gráfico
    plt.show()

# 7. Probar la función con un número grande de lanzamientos
analizar_lanzamientos_dado(10000)
```

### **Ejercicio 2: Pipeline de Manipulación de Arrays (Reshape, Transpose, Concatenate)**

**Instrucciones:**

1.  **Importa `numpy` como `np`.**

2.  **Datos Iniciales**: Imagina que tienes datos económicos para dos trimestres. Crea dos arrays 1D:
    * `trimestre_1`: Un array con los números del 10 al 15.
    * `trimestre_2`: Un array con los números del 16 al 21.

3.  **Paso 1: Remodelar (Reshape)**
    * Los datos de cada trimestre corresponden a 3 meses, con 2 mediciones por mes. Remodela ambos arrays a una forma de **(3, 2)** (3 meses, 2 mediciones). Llama a los nuevos arrays `matriz_q1` y `matriz_q2`.
    * Imprime ambas matrices.

4.  **Paso 2: Transponer (Transpose)**
    * Para un reporte, necesitas que las mediciones estén en las filas y los meses en las columnas. Transpón la `matriz_q1` usando el atributo `.T` y guárdala en `matriz_q1_transpuesta`.
    * Imprime la matriz transpuesta para ver el cambio de forma a (2, 3).

5.  **Paso 3: Concatenar (Concatenate)**
    * Ahora, crea un único dataset que contenga los datos de ambos trimestres uno al lado del otro.
    * Usa `np.concatenate()` para unir `matriz_q1` y `matriz_q2` de forma **horizontal (`axis=1`)**. Esto creará una matriz final de 3x4. Llama al resultado `datos_semestre`.
    * Imprime la matriz `datos_semestre` final.

```python
# 1. Importar numpy
import numpy as np

# 2. Datos Iniciales
trimestre_1 = np.arange(10, 16)
trimestre_2 = np.arange(16, 22)
print("--- Datos Crudos Iniciales ---")
print(f"Trimestre 1: {trimestre_1}")
print(f"Trimestre 2: {trimestre_2}\n")


# 3. Paso 1: Remodelar (Reshape)
print("--- Paso 1: Datos Remodelados a (3, 2) ---")
matriz_q1 = trimestre_1.reshape(3, 2)
matriz_q2 = trimestre_2.reshape(3, 2)
print("Matriz Trimestre 1:")
print(matriz_q1)
print("\nMatriz Trimestre 2:")
print(matriz_q2)
print("\n" + "="*40 + "\n")


# 4. Paso 2: Transponer (Transpose)
print("--- Paso 2: Transponiendo la Matriz del Trimestre 1 ---")
matriz_q1_transpuesta = matriz_q1.T
print("Matriz T1 Transpuesta (2, 3):")
print(matriz_q1_transpuesta)
print("\n" + "="*40 + "\n")


# 5. Paso 3: Concatenar (Concatenate)
print("--- Paso 3: Uniendo los datos de ambos trimestres horizontalmente ---")
# Une las matrices originales de 3x2 horizontalmente -> (3, 2+2) = (3, 4)
datos_semestre = np.concatenate((matriz_q1, matriz_q2), axis=1)
print("Dataset final del semestre (3, 4):")
print(datos_semestre)
```

2.  **Define los Parámetros de la Simulación**: Crea variables para los parámetros que usarás:
    * `media_teorica = 100` (el centro de nuestra distribución).
    * `desviacion_estandar = 15`.
    * `numero_de_muestras = 10000` (usamos un número grande para que la forma de la distribución sea clara).

3.  **Genera los Datos**: Utiliza `np.random.normal()` con los parámetros `loc` (media), `scale` (desviación estándar) y `size` para crear un array con los datos simulados.

4.  **Verificación Numérica**:
    * Calcula la **media real** del array que acabas de generar usando `np.mean()`.
    * Calcula la **mediana real** usando `np.median()`.
    * Imprime un resumen que compare la `media_teorica` con la media y la mediana que calculaste. (Deberían ser muy similares).

5.  **Verificación Visual con un Histograma**:
    * Usa `plt.hist()` para crear un histograma de tus datos simulados. Elige un número de `bins` adecuado, como 50.

```python
# 1. Importar las librerías
import numpy as np
import matplotlib.pyplot as plt

# 2. Definir los Parámetros de la Simulación
media_teorica = 100
desviacion_estandar = 15
numero_de_muestras = 10000

print("--- Simulación de Datos de una Distribución Normal ---")
print(f"Parámetros: Media={media_teorica}, Desv. Est.={desviacion_estandar}, Muestras={numero_de_muestras:,}\n")

# 3. Generar los Datos
datos_simulados = np.random.normal(loc=media_teorica, scale=desviacion_estandar, size=numero_de_muestras)

# 4. Verificación Numérica
media_real = np.mean(datos_simulados)
mediana_real = np.median(datos_simulados)

print("--- Verificación Numérica ---")
print(f"Media Teórica:   {media_teorica}")
print(f"Media Real:      {media_real:.4f}")
print(f"Mediana Real:    {mediana_real:.4f}\n")

# 5. Verificación Visual con un Histograma
print("--- Verificación Visual ---")
print("Generando histograma...")
plt.hist(datos_simulados, bins=50, density=True, color='cyan', edgecolor='black', alpha=0.7, label='Distribución Simulada')

# Añadir una línea vertical para marcar la media teórica
plt.axvline(media_teorica, color='red', linestyle='--', linewidth=2, label=f'Media Teórica = {media_teorica}')

# Personalización del gráfico
plt.title('Verificación de Simulación de Distribución Normal')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, alpha=0.5)

# 6. Mostrar el gráfico
plt.show()
```

### **Ejercicio 5: Unión de Datasets con `np.concatenate` (Apilado Vertical vs. Horizontal)**

**Instrucciones:**

1.  **Importa `numpy` como `np`.**

2.  **Simulación de Datos**: Imagina que tienes los datos de ventas trimestrales (en miles de USD) para 3 productos durante dos años consecutivos.
    * Crea una matriz de 3x4 llamada `ventas_anio1`. Puedes llenarla con números enteros aleatorios entre 100 y 500.
    * Crea una segunda matriz de 3x4 llamada `ventas_anio2` con datos similares.
    * Imprime ambas matrices iniciales con una descripción.

3.  **Apilado Vertical (`axis=0`)**:
    * Utiliza `np.concatenate()` para unir las dos matrices de forma que las filas de `ventas_anio2` queden debajo de las de `ventas_anio1`. Esto crea un historial de ventas unificado.
    * El resultado debe ser una matriz de 6x4.
    * Imprime esta nueva matriz con el título "Historial de Ventas (Apilado Vertical)".

4.  **Apilado Horizontal (`axis=1`)**:
    * Ahora, utiliza `np.concatenate()` de nuevo, pero esta vez une las matrices de forma que las columnas de `ventas_anio2` queden al lado de las de `ventas_anio1`. Esto sirve para comparar los años lado a lado.
    * El resultado debe ser una matriz de 3x8.
    * Imprime esta nueva matriz con el título "Comparativa Anual (Apilado Horizontal)".

```python
# 1. Importar numpy
import numpy as np

# Para que los resultados aleatorios sean siempre los mismos y el ejemplo sea reproducible
np.random.seed(42)

# 2. Simulación de Datos
print("--- Datos Iniciales ---")
ventas_anio1 = np.random.randint(100, 501, size=(3, 4))
ventas_anio2 = np.random.randint(100, 501, size=(3, 4))

print("Ventas del Año 1 (3 productos, 4 trimestres):")
print(ventas_anio1)
print("\nVentas del Año 2:")
print(ventas_anio2)
print("\n" + "="*50 + "\n")


# 3. Apilado Vertical (axis=0)
print("--- 3. Historial de Ventas (Apilado Vertical, axis=0) ---")
# Une los arrays apilando las filas. El resultado es (3+3)x4 -> 6x4
historial_ventas = np.concatenate((ventas_anio1, ventas_anio2), axis=0)
print("Forma del array resultante:", historial_ventas.shape)
print(historial_ventas)
print("\n" + "="*50 + "\n")


# 4. Apilado Horizontal (axis=1)
print("--- 4. Comparativa Anual (Apilado Horizontal, axis=1) ---")
# Une los arrays añadiendo las columnas. El resultado es 3x(4+4) -> 3x8
comparativa_anual = np.concatenate((ventas_anio1, ventas_anio2), axis=1)
print("Forma del array resultante:", comparativa_anual.shape)
print(comparativa_anual)
```

### **Ejercicio 8: Simulación y Visualización de un Paseo Aleatorio (Random Walk)**

**Instrucciones:**

1.  **Importa `numpy` y `matplotlib.pyplot`** con sus alias convencionales.

2.  **Define los Parámetros de la Simulación**:
    * `precio_inicial = 100` (el precio de partida del activo).
    * `num_dias = 252` (aproximadamente un año de días de mercado).

3.  **Genera los Pasos Aleatorios**: Simula los cambios de precio diarios. Para hacerlo más realista, usaremos una distribución normal.
    * Crea un array llamado `cambios_diarios` con `num_dias` puntos, usando `np.random.normal(loc=0.1, scale=1.5, size=num_dias)`. Esto simula que el precio tiende a subir un poco cada día (`loc=0.1`) pero con bastante volatilidad (`scale=1.5`).

4.  **Construye la Trayectoria de Precios**:
    * Inicializa una lista de Python llamada `trayectoria_precios` que contenga únicamente el `precio_inicial`.
    * Usa un **bucle `for`** para recorrer cada `cambio` en el array `cambios_diarios`.
    * En cada iteración, calcula el nuevo precio (el último precio de la `trayectoria_precios` más el `cambio` actual) y añádelo a la lista `trayectoria_precios` con `.append()`.

5.  **Visualización con un Gráfico de Líneas**:
    * Usa `plt.plot()` para graficar la `trayectoria_precios` que has construido.
    * **Personaliza el gráfico**: Añade un título (ej. "Simulación de Paseo Aleatorio de un Precio"), y etiquetas para los ejes X ("Días de Mercado") e Y ("Precio del Activo"). Añade una cuadrícula (`plt.grid(True)`) para facilitar la lectura.

6.  **Muestra el gráfico** final con `plt.show()`.

```python
# 1. Importar las librerías
import numpy as np
import matplotlib.pyplot as plt

# 2. Definir los Parámetros
precio_inicial = 100
num_dias = 252

# Para reproducibilidad
np.random.seed(101)

print(f"--- Simulación de un Paseo Aleatorio para {num_dias} días ---")
print(f"Precio de partida: ${precio_inicial}\n")

# 3. Generar los Pasos Aleatorios
# Simulamos pequeños cambios diarios con una tendencia ligeramente positiva
cambios_diarios = np.random.normal(loc=0.1, scale=1.5, size=num_dias)

# 4. Construir la Trayectoria de Precios
trayectoria_precios = [precio_inicial] # La lista empieza con el precio inicial

# Bucle para construir la trayectoria
for cambio in cambios_diarios:
    # El nuevo precio es el último precio de la lista + el cambio de hoy
    nuevo_precio = trayectoria_precios[-1] + cambio
    trayectoria_precios.append(nuevo_precio)

# 5. Visualización con un Gráfico de Líneas
print("Generando gráfico de la trayectoria...")
plt.figure(figsize=(10, 6)) # Opcional: hace el gráfico más grande y legible
plt.plot(trayectoria_precios, color='navy')

# Personalización
plt.title('Simulación de Paseo Aleatorio de un Precio')
plt.xlabel('Días de Mercado')
plt.ylabel('Precio del Activo ($)')
plt.grid(True, linestyle='--', alpha=0.6)

# 6. Mostrar el gráfico
plt.show()
```

    * `num_observaciones = 150` (el número de personas en nuestra muestra).
    * `ingreso_base = 10000` (el ingreso para alguien con 0 años de estudio).
    * `retorno_por_anio_estudio = 2500` (cuánto aumenta el ingreso por cada año extra de estudio).

3.  **Simulación de Datos**:
    * **Variable Independiente (X)**: Crea un array llamado `anios_estudio` con `num_observaciones` números enteros aleatorios entre 8 y 22.
    * **Variable Dependiente (Y)**: Crea un array llamado `ingreso_estimado`. Su valor debe ser el `ingreso_base` más (`anios_estudio` * `retorno_por_anio_estudio`), más un "ruido" aleatorio para hacerlo más realista. Añade este ruido usando `np.random.normal(loc=0, scale=5000, size=num_observaciones)`.

4.  **Visualización con `seaborn.regplot`**:
    * Utiliza la función `sns.regplot()` para crear el gráfico. Pasa `anios_estudio` al eje `x` e `ingreso_estimado` al eje `y`.
    * Esta función dibujará automáticamente el gráfico de dispersión y le superpondrá la línea de regresión lineal con su intervalo de confianza.

    * Usa `plt.title()`, `plt.xlabel()` y `plt.ylabel()` para dar contexto al gráfico.

```python
# 1. Importar las librerías
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Definir los Parámetros
num_observaciones = 150
ingreso_base = 10000
retorno_por_anio_estudio = 2500

# Para que la simulación sea reproducible
np.random.seed(42)

print("--- Simulación de la Relación entre Educación e Ingreso ---")

# 3. Simulación de Datos
# Variable X: Años de estudio (entre 8 y 22)
anios_estudio = np.random.randint(8, 23, size=num_observaciones)

# Variable Y: Ingreso = (base + efecto_educacion) + ruido_aleatorio
ruido = np.random.normal(loc=0, scale=8000, size=num_observaciones)
ingreso_estimado = ingreso_base + (anios_estudio * retorno_por_anio_estudio) + ruido

print("Simulación completada. Generando visualización...\n")

# 4. Visualización con seaborn.regplot
plt.figure(figsize=(10, 6)) # Opcional: crea una figura más grande
sns.regplot(x=anios_estudio, y=ingreso_estimado,
            scatter_kws={'alpha':0.6, 'color':'blue'}, # Estilo para los puntos
            line_kws={'color':'red', 'linewidth':3})   # Estilo para la línea

# 5. Personalizar y mostrar el gráfico
plt.title('Relación Simulada entre Años de Estudio e Ingreso', fontsize=16)
plt.xlabel('Años de Estudio', fontsize=12)
plt.ylabel('Ingreso Anual Estimado ($)', fontsize=12)
plt.grid(True)
plt.show()
```

### **Ejercicio 10: Visualización Comparativa de Múltiples Funciones Matemáticas**

2.  **Generación del Eje X**: Crea un array `x` que contenga 100 puntos distribuidos uniformemente entre -10 y 10. La función `np.linspace()` es ideal para esto.

3.  **Generación de los Ejes Y**: Usando **operaciones vectorizadas** sobre el array `x`, crea tres arrays `y` diferentes:
    * `y1 = x` (Función Lineal)
    * `y2 = x**2` (Función Cuadrática)
    * `y3 = x**3` (Función Cúbica)

4.  **Creación y Personalización del Gráfico**:
    * Opcional: Usa `plt.figure(figsize=(10, 7))` para crear un lienzo de un tamaño más grande y legible.
    * Llama a `plt.plot()` **tres veces**, una para cada par `(x, y)`.
    * En cada llamada a `plt.plot()`, personaliza la línea usando los siguientes argumentos:
        * **Para y1**: `color='blue'`, `linestyle='--'`, `label='y = x'`
        * **Para y2**: `color='red'`, `linestyle='-'`, `label='y = x²'`
        * **Para y3**: `color='green'`, `linestyle=':'`, `label='y = x³'`

5.  **Añade los Toques Finales**:
    * Añade un título general, etiquetas para los ejes X e Y, y una cuadrícula (`plt.grid(True)`).

```python
# 1. Importar las librerías
import numpy as np
import matplotlib.pyplot as plt

# 2. Generación del Eje X
# 100 puntos entre -10 y 10 para curvas suaves
x = np.linspace(-10, 10, 100)

# 3. Generación de los Ejes Y
y1 = x           # Lineal
y2 = x**2        # Cuadrática
y3 = x**3        # Cúbica

# 4. Creación y Personalización del Gráfico
print("--- Generando Gráfico Comparativo de Funciones ---")
plt.figure(figsize=(10, 7)) # Crea una figura de 10x7 pulgadas

# Graficar cada función con su propio estilo y etiqueta
plt.plot(x, y1, color='blue', linestyle='--', label='y = x (Lineal)')
plt.plot(x, y2, color='red', linestyle='-', label='y = x² (Cuadrática)')
plt.plot(x, y3, color='green', linestyle=':', label='y = x³ (Cúbica)')

# 5. Añade los Toques Finales
plt.title('Comparación de Funciones Polinomiales', fontsize=16)
plt.xlabel('Eje X', fontsize=12)
plt.ylabel('Eje Y', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5) # Línea horizontal en y=0
plt.axvline(0, color='black', linewidth=0.5) # Línea vertical en x=0

# La leyenda usa las etiquetas 'label' definidas en cada plt.plot()
plt.legend(fontsize=12)

# 6. Mostrar el gráfico
plt.show()
```

```python
# 1. Importar las librerías
import numpy as np
import matplotlib.pyplot as plt

# Para que los resultados aleatorios sean consistentes
np.random.seed(42)

# 2. Simulación de un Dataset de Productos
num_productos = 150

# Variable X: Precio
precios = np.random.uniform(low=20, high=150, size=num_productos)

# Variable Y: Unidades vendidas (con relación negativa al precio + ruido)
ruido_ventas = np.random.normal(loc=0, scale=80, size=num_productos)
unidades_vendidas = 1200 - (precios * 6) + ruido_ventas

# Variable para el Color: Margen de beneficio
margen_beneficio = np.random.uniform(low=0.05, high=0.40, size=num_productos)

# Variable para el Tamaño: Popularidad (ej. menciones en redes sociales)
popularidad = np.random.uniform(low=10, high=500, size=num_productos)


# 3. Creación del Gráfico de Dispersión Avanzado
print("--- Generando Gráfico de Dispersión Multidimensional ---")
plt.figure(figsize=(12, 8))

scatter = plt.scatter(x=precios,
                      y=unidades_vendidas,
                      c=margen_beneficio,
                      s=popularidad,
                      cmap='viridis',
                      alpha=0.7,
                      edgecolors='black',
                      linewidth=0.5)

# 4. Personalización y Contexto
# Añadir la barra de color como leyenda para el margen de beneficio
cbar = plt.colorbar(scatter)
cbar.set_label('Margen de Beneficio (%)', fontsize=12)

# Añadir títulos y etiquetas
plt.title('Análisis de Productos: Precio, Ventas, Margen y Popularidad', fontsize=16)
plt.xlabel('Precio de Venta ($)', fontsize=12)
plt.ylabel('Unidades Vendidas', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)


# 5. Mostrar el gráfico
plt.show()
```

### **Ejercicio 12: Comparación Visual de Distribuciones (Uniforme vs. Normal)**

**Instrucciones:**

1.  **Importa `numpy` y `matplotlib.pyplot`** con sus alias `np` y `plt`.

2.  **Define los Parámetros**: Establece una variable `cantidad_datos = 20000` para generar una muestra grande y clara.

3.  **Generación de Datos**:
    * Crea un array llamado `datos_uniformes` con `cantidad_datos`, usando `np.random.uniform(low=0, high=100)`.
    * Crea un segundo array llamado `datos_normales`. Para que sea comparable, usa una media de 50 (el punto medio del rango uniforme) y una desviación estándar de 15. Usa `np.random.normal(loc=50, scale=15, size=cantidad_datos)`.

4.  **Creación de Subplots**:
    * Utiliza `plt.subplots(1, 2, figsize=(14, 6))` para crear una figura que contenga dos gráficos (1 fila, 2 columnas). Esto devuelve una figura (`fig`) y un array de ejes (`axes`). `axes[0]` será el gráfico de la izquierda y `axes[1]` el de la derecha.

5.  **Graficar en cada Subplot**:
    * **En el primer eje (`axes[0]`)**: Dibuja el histograma de `datos_uniformes`. Usa la sintaxis `axes[0].hist(...)` y establece su título con `axes[0].set_title('Distribución Uniforme')`.
    * **En el segundo eje (`axes[1]`)**: Dibuja el histograma de `datos_normales`. Usa la sintaxis `axes[1].hist(...)` y establece su título con `axes[1].set_title('Distribución Normal')`.

7.  **Muestra el gráfico** con `plt.show()`.

```python
# 1. Importar las librerías
import numpy as np
import matplotlib.pyplot as plt

# 2. Definir los Parámetros
cantidad_datos = 20000

# 3. Generación de Datos
print("--- Generando datos para ambas distribuciones ---")
# Distribución Uniforme entre 0 y 100
datos_uniformes = np.random.uniform(low=0, high=100, size=cantidad_datos)

# Distribución Normal centrada en 50 para que sea comparable
datos_normales = np.random.normal(loc=50, scale=15, size=cantidad_datos)
print("Datos generados. Creando visualización...\n")

# 4. Creación de Subplots
# fig es la figura completa, axes es un array con los dos subplots [ax1, ax2]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# 5. Graficar en cada Subplot
# Gráfico de la izquierda: Distribución Uniforme
axes[0].hist(datos_uniformes, bins=50, color='darkorange', edgecolor='black', alpha=0.7)
axes[0].set_title('Distribución Uniforme', fontsize=14)
axes[0].set_xlabel('Valor')
axes[0].set_ylabel('Frecuencia')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico de la derecha: Distribución Normal
axes[1].hist(datos_normales, bins=50, color='dodgerblue', edgecolor='black', alpha=0.7)
axes[1].set_title('Distribución Normal', fontsize=14)
axes[1].set_xlabel('Valor')
# No es necesario poner la etiqueta Y en el segundo si comparten la misma escala

# 6. Añade un Título General
fig.suptitle('Comparación de Distribuciones de Probabilidad', fontsize=18)

# Ajusta el layout para que no se solapen los títulos
plt.tight_layout(rect=[0, 0, 1, 0.96])

# 7. Muestra el gráfico
plt.show()
```

### **Ejercicio 13: Gráfico Comparativo de Series de Tiempo con Seaborn y Matplotlib**

**Instrucciones:**

1.  **Importa las librerías** `numpy` como `np`, `matplotlib.pyplot` como `plt`, y `seaborn` como `sns`.

2.  **Configura el Estilo de Seaborn**: Aplica un tema visualmente atractivo para todos tus gráficos usando `sns.set_theme(style='whitegrid')`.

3.  **Simulación de Datos de Activos**:
    * Define los parámetros: `num_dias = 120`, `precio_inicial = 75`.
    * **Activo Estable**: Genera un array de cambios diarios (`pasos_estables`) usando `np.random.normal(loc=0.05, scale=0.5, size=num_dias)`.
    * **Activo Volátil**: Genera otro array de cambios (`pasos_volatiles`) usando `np.random.normal(loc=0.05, scale=1.5, size=num_dias)`.
    * **Construye las trayectorias**: Usa `np.cumsum()` para calcular la trayectoria de precios de cada activo a partir del `precio_inicial` y sus respectivos pasos.

4.  **Creación y Personalización del Gráfico**:
    * Crea una figura de un tamaño adecuado (ej. `12x7`) con `plt.figure()`.
    * Usa `sns.lineplot()` para graficar la trayectoria del **activo estable**. Asígnale un `label`.
    * En el mismo plano, usa `sns.lineplot()` de nuevo para graficar la trayectoria del **activo volátil**, con otro `label`.

5.  **Añade Toques Profesionales**:
    * Establece un **título principal** y **etiquetas para los ejes** X e Y.
    * Coloca la **leyenda** en la mejor ubicación posible usando `plt.legend(loc='best')`.
    * Añade una línea horizontal en el precio inicial para tener una referencia visual, usando `plt.axhline()`.

6.  **Guarda y Muestra el Gráfico**:
    * **Guarda la figura** en un archivo de alta resolución llamado `"evolucion_activos.png"` usando `plt.savefig()`.
    * Finalmente, **muestra el gráfico** en pantalla con `plt.show()`.

```python
# 1. Importar las librerías
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Configurar el Estilo de Seaborn
sns.set_theme(style="whitegrid")

# Para reproducibilidad de la simulación
np.random.seed(123)

# 3. Simulación de Datos de Activos
num_dias = 120
precio_inicial = 75

# Simulación de los cambios diarios
pasos_estables = np.random.normal(loc=0.08, scale=0.5, size=num_dias)
pasos_volatiles = np.random.normal(loc=0.08, scale=1.8, size=num_dias)

# Construcción de las trayectorias usando np.cumsum()
# cumsum() calcula la suma acumulada de los pasos
trayectoria_estable = precio_inicial + np.cumsum(pasos_estables)
trayectoria_volatil = precio_inicial + np.cumsum(pasos_volatiles)

# 4. Creación y Personalización del Gráfico
print("--- Generando gráfico final de comparación de activos ---")
plt.figure(figsize=(12, 7))

# Graficar ambas series de tiempo con Seaborn
sns.lineplot(x=range(num_dias), y=trayectoria_estable, label='Activo Estable', color='royalblue', linewidth=2)
sns.lineplot(x=range(num_dias), y=trayectoria_volatil, label='Activo Volátil', color='firebrick', linewidth=2)

# 5. Añade Toques Profesionales
plt.title('Simulación de la Evolución de Dos Activos Financieros', fontsize=18)
plt.xlabel('Días de Mercado', fontsize=12)
plt.ylabel('Precio de Cierre ($)', fontsize=12)
plt.legend(loc='best', fontsize=12)
plt.axhline(y=precio_inicial, color='grey', linestyle='--', label=f'Precio Inicial = ${precio_inicial}')

# 6. Guarda y Muestra el Gráfico
# Guardar ANTES de mostrar
plt.savefig('evolucion_activos.png', dpi=300)

print("Gráfico guardado como 'evolucion_activos.png'. Mostrando visualización...")
plt.show()
```

### **Ejercicio 14: Visualización de la Ley de los Grandes Números (Simulación de Moneda)**

**Instrucciones:**

1.  **Importa `numpy` y `matplotlib.pyplot`** con sus alias `np` y `plt`.

2.  **Define los Parámetros**: Establece una variable `num_lanzamientos = 5000` para tener una simulación lo suficientemente larga.

3.  **Simulación Vectorizada**: Usa `np.random.choice()` para generar un array con el resultado de todos los `num_lanzamientos` de una moneda (`'Cara'` o `'Cruz'`).

4.  **Cálculo de la Convergencia**:
    * Crea un array booleano que sea `True` donde el resultado fue `'Cara'` y `False` en caso contrario.
    * Utiliza `np.cumsum()` sobre este array booleano para obtener la suma acumulada de caras después de cada lanzamiento. `np.cumsum()` es la "suma acumulada", que para `[0, 1, 0, 1]` devolvería `[0, 1, 1, 2]`.
    * Crea un segundo array que represente el número del lanzamiento actual (1, 2, 3, ...). `np.arange(1, num_lanzamientos + 1)` es perfecto para esto.
    * Divide el array de sumas acumuladas por el array de número de lanzamientos para obtener la **proporción acumulada** en cada punto.

5.  **Visualización del Proceso**:
    * Utiliza `plt.plot()` para crear un gráfico de línea. El eje X será el número de lanzamientos y el eje Y será la proporción acumulada.
    * Añade una **línea de referencia horizontal** en `y=0.5` para representar la probabilidad teórica. Usa `plt.axhline()`.
    * **Personaliza el gráfico**: Añade un título, etiquetas para los ejes, y ajusta los límites del eje Y (por ejemplo, de 0 a 1) para una mejor visualización usando `plt.ylim()`.

6.  **Muestra el gráfico** y observa cómo la proporción converge hacia 0.5.

```python
# 1. Importar las librerías
import numpy as np
import matplotlib.pyplot as plt

# 2. Definir los Parámetros
num_lanzamientos = 5000

print(f"--- Simulación de {num_lanzamientos:,} lanzamientos de una moneda ---")
print("Visualizando la Ley de los Grandes Números...")

# 3. Simulación Vectorizada
lanzamientos = np.random.choice(['Cara', 'Cruz'], size=num_lanzamientos)

# 4. Cálculo de la Convergencia
# Convertir a 1s (Cara) y 0s (Cruz) para cálculos numéricos
es_cara = (lanzamientos == 'Cara')

# Calcular la suma acumulada de caras en cada paso
suma_acumulada_caras = np.cumsum(es_cara)

# Crear un array con el número de lanzamiento (1, 2, 3, ...)
indices_lanzamiento = np.arange(1, num_lanzamientos + 1)

# Calcular la proporción acumulada en cada paso
proporcion_acumulada = suma_acumulada_caras / indices_lanzamiento


# 5. Visualización del Proceso
plt.figure(figsize=(12, 7))
plt.plot(indices_lanzamiento, proporcion_acumulada, label='Proporción Observada', color='dodgerblue')

# Añadir la línea de referencia teórica
plt.axhline(y=0.5, color='red', linestyle='--', linewidth=2, label='Probabilidad Teórica (0.5)')

# Personalización del gráfico
plt.title('Convergencia de la Proporción de Caras (Ley de los Grandes Números)', fontsize=16)
plt.xlabel('Número de Lanzamientos', fontsize=12)
plt.ylabel('Proporción Acumulada de Caras', fontsize=12)
plt.xlim(0, num_lanzamientos)
plt.ylim(0, 1) # Fijar el eje Y entre 0 y 1
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# 6. Muestra el gráfico
plt.show()
```
