9.Genera 100 números entre 0 y 10 y haz un histograma con el formato de tu preferencia.



```python
import matplotlib.pyplot as plt

numeros = np.random.randint(0, 11, size=100)
plt.hist(numeros, bins=11, color= 'orange', edgecolor='black')
plt.title("Histograma de números aleatorios")
plt.show()
```

10.Gráfica los puntos x =[0,1,2,3,4], y =[0,1,4,9,16] con un gráfico de línea.

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y)
plt.title("Gráfico de Línea: y = x²")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

11.Genera 100 puntos aleatorios para x e y entre 0 y 1 y grafícalos con un gráfico de dispersión.


    

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, alpha=0.6)
plt.title("Dispersión de puntos aleatorios")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

13.Gráfica y1=x y y2=x² en el mismo plano con un gráfico de línea.

```python
x = np.arange(0, 11)
y1 = x
y2 = x**2

plt.plot(x, y1, label="y = x")
plt.plot(x, y2, label="y = x²")
plt.title("Comparación de funciones")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

14.Simula 100 lanzamientos de una moneda y calcula la proporción de caras.

```python
lanzamientos = np.random.choice(['cara', 'cruz'], size=100)
proporcion_caras = np.mean(lanzamientos == 'cara')
print("Proporción de caras:", proporcion_caras)
```
