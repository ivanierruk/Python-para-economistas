1.Crea una Serie que contenga los nombres de 5 países.


```python
import pandas as pd

paises = ['México', 'Brasil', 'Argentina', 'Chile', 'Colombia']
serie_paises = pd.Series(paises)
print(serie_paises)
```

4.Crea una Serie con nombres de frutas (manzana, fresa, uva) y accede al valor asociado con "manzana".

```python
frutas = pd.Series([3, 5, 2], index=['manzana', 'fresa', 'uva'])
print(frutas['manzana'])
```

5.Dadas dos Series que representan ventas en enero y febrero, suma ambas para obtener las ventas totales.



*   Enero: 100, 150, 200
*   Febrero: 120, 130, 180



```python
enero = pd.Series([100, 150, 200])
febrero = pd.Series([120, 130, 180])
total = enero + febrero
print(total)
```

6. Usa una Serie de edades y determina cuáles son mayores de edad (≥18).

```python
edades = pd.Series([15, 20, 17, 25])
mayores = edades >= 18
print(mayores)
```

8.Convierte un array NumPy en un DataFrame de 2 columnas.

```python

import numpy as np

array = np.array([[1, 2], [3, 4], [5, 6]])
df_array = pd.DataFrame(array, columns=['Columna1', 'Columna2'])
print(df_array)
```

12.Crea un DataFrame con índice personalizado y usa .loc[] para seleccionar una fila.



```python
df = pd.DataFrame({'Nombre': ['Ana', 'Luis', 'Sofía'], 'Edad': [20, 25, 30]}, index=['a', 'b', 'c'])
print(df.loc['b'])
```
