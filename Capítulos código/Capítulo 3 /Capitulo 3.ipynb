# Capitulo 3. Bucles

Los **bucles** son una herramienta esencial en programación que permite ejecutar un bloque de código repetidamente. Por ejemplo, si tenemos una lista de Python con 5 elementos “lista = [1, 2, 3, 4, 5]", los bucles nos permiten iterar, es decir, recorrer o repetir un proceso sobre cada elemento de la lista. En este ejemplo de la “lista”, un bucle recorrería cada número de la lista y ejecutaríamos alguna acción sobre cada elemento, por ejemplo, imprimirlo. Python ofrece dos tipos principales de bucles, el bucle “for” y “while”:
- En un bucle “for”, iterar implica procesar cada elemento de una secuencia, como recorrer una lista de números para imprimir cada uno.
- En un bucle while, iterar implica repetir el código hasta que la condición deje de ser verdadera.

#### 3.1 Bucles for: Iteración sobre Secuencias

El bucle "for" en Python es una forma concisa de iterar sobre los elementos de una secuencia. La sintaxis básica es la palabra clave “for” seguida de una variable (nosotros asignamos el nombre que queramos a esa variable), lposteriormente utilizamos la palabra clave “in”, y por último el objeto sobre el cual queremos que itere, en el siguiente ejemplo usamos una lista para iterar sobre cada uno de los elementos e imprimirlos.   

```python
lista = [1,2,3,4,5]
for numero in lista:
  # Código a jectuar para cada elemento
    print (numero)
```

El bucle for recorrerá cada elemento dentro de la secuencia especificada y ejecutará el bloque de código indentado para cada uno de estos elementos. En el caso de las listas, el bucle “for” asigna cada elemento de la lista a la variable de iteración (“numero” en el ejemplo mostrado anteriormente).

Cuando se usa un bucle “for” para iterar una lista, a veces necesitamos saber no solo el valor de cada elemento, sino también su posición (o índice) en la lista, la función “enumerate()” en Python, ayuda a obtener ambos al mismo tiempo: el índice (un número que indica la posición) y el valor del elemento.

```python
productos = ["Cobre", "Petróleo", "Maíz"]

for indice, producto in enumerate(productos):
    print(f"Índice: {indice}, Producto: {producto}")
```

#### 3.1.1 La Función range()

A menudo, es necesario iterar un cierto número de veces. La función incorporada range() es muy útil para generar una secuencia de números. Puede utilizarse de tres formas:
- range(fin): Genera una secuencia de números desde 0 hasta fin - 1.
- range(inicio, fin): Genera una secuencia de números desde inicio hasta fin - 1.
- range(inicio, fin, paso): Genera una secuencia de números desde inicio hasta fin - 1, con un incremento de paso.

```python
# Iterar 5 veces (de 0 a 4)
for i in range(5):
   print(i)
# Iterar de 2 a 5 (2, 3, 4,5)
for j in range(2, 6):
    print("El valor actual es:", j)
# Iterar de 0 a 10 con un paso de 2 (0, 2, 4, 6, 8)
for k in range(0, 11, 2):
    print("Número par:", k)
```

La función range() es comúnmente utilizada en bucles for para controlar el número de iteraciones.

#### 3.1.2 Bucles Anidados

Es posible colocar un bucle dentro de otro bucle, lo que se conoce como bucles anidados. Esto puede ser útil para iterar sobre estructuras de datos multidimensionales, como listas de listas (que pueden representar matrices).

```python
matriz_precios = [[10, 12, 11], [15, 16, 14], [13, 14, 15]]

for fila in matriz_precios:
    print("Fila:", fila)
    for precio in fila:
        print("Precio:", precio)
```

En el código anterior, el "for” exterior itera sobre cada fila de la “matriz_precios”, y el “for” interior itera sobre cada precio dentro de la fila actual. Los bucles anidados pueden ser útiles para realizar operaciones en cada elemento de una matriz o para generar combinaciones.

#### 3.2 Bucles while: Iteración Condicional

El bucle “while” ejecuta un bloque de código repetidamente **mientras una condición especificada se mantenga verdadera**. A diferencia del bucle “for”, que itera sobre una secuencia finita, el “while” continúa su ejecución de forma indefinida hasta que la condición de control se evalúa como “False”. Su sintaxis básica es la palabra “while” segida de una condición.  

`while condicion:` Es el código a ejecutar mientras esa condición sea verdadera

Es crucial asegurarse de que la condición eventualmente se vuelva falsa dentro del bucle para evitar un **bucle infinito**, donde el código se ejecuta indefinidamente.

```python
contador = 0

while contador < 5:
    print("El contador es:", contador)
    contador += 1 # Incrementamos el contador en cada iteración
```

En este ejemplo, el bucle “while” se ejecutará mientras la variable “contador” sea menor que 5. En cada iteración, se imprime el valor del contador y luego se incrementa usando la expresión “+=”. Eventualmente, la condición (contador < 5) se vuelve falsa y el bucle termina. La expresión “+=” puede parecer confusa, pero en realidad lo que estamos haciendo es sumar el valor “1” a la variable contador (contador = contador + 1), en Python utilizamos una expresión más concisa (+=) para representar esta operación.

Definir correctamente la condición de terminación es fundamental para evitar bucles infinitos en los bucles “while”. La condición debe basarse en una o más variables que se modifican dentro del bucle de tal manera que eventualmente la condición se evalúe como falsa.

```python
# Definir las variables
tasa_crecimiento = 0.01
año = 2023
pib = 1000

# Inicial el bucle “while”
while (tasa_crecimiento < 0.05 and año < 2030):
    pib *= (1 + tasa_crecimiento)
    print("PIB en el año", año, ":", pib)
    print("Tasa de crecimiento:", tasa_crecimiento)

    tasa_crecimiento += 0.005 # Simulación de un aumento gradual en la tasa
    año += 1
```

En esta simulación simplificada, el bucle while continúa mientras la tasa de crecimiento sea menor que 0.05 y el año sea anterior a 2030. En cada iteración, se actualiza el PIB, se imprime la información y se modifican tanto la tasa de crecimiento como el año, lo que eventualmente llevará a que ambas condiciones sean falsas y el bucle termine.

#### 3.3 List Comprehensions

Las “list comprehensions” son una característica poderosa y concisa de Python que permite iterar en objetos como (listas, rangos, diccionarios, etc.) de forma más compacta que usando bucles tradicionales. Con una sola línea de código, es posible generar una nueva lista aplicando una operación o condición a cada elemento de un iterable. Son ideales para simplificar el código y hacerlo más legible, especialmente cuando se requiere transformar o filtrar datos.

La sintaxis básica es: “[expresión for elemento in iterable]”, el siguiente código muestra a manera de ejemplo el uso de una “list comprehension”.

```python
import numpy as np

lista= [1, 2, 3, 4, 5]
array_np = np.array(lista)
nueva_lista = [x * 2.5 for x in lista]
nuevo_array_np = [x * 2.5 for x in array_np]

print(nueva_lista, nuevo_array_np, sep="\n")
```

También podemos colocar condiciones `if`,  `else` en las `list comprehensions`, en el siguiente ejemplo, la parte `if` condición es opcional y se utiliza para filtrar los elementos del iterable

```python
lista= [x for x in range(6) if x > 3]

print(lista)
```

#### 3.4 Ejemplos de simulación iterativa  

La programación en Python nos permite simular fenómenos para entender sus propiedades. Un método muy potente es la simulación de Montecarlo, que consiste en repetir un experimento aleatorio miles de veces para estimar una probabilidad o un valor esperado.

#### **Ejemplo 1.**
Aunque podríamos usar un bucle “for” para simular un lanzamiento de moneda 10,000 veces, es mucho más eficiente usar el poder de **vectorización** de NumPy para realizar todas las simulaciones en un solo paso. Posteriormente, usaremos una forma compacta de iteración (list comprehension) para visualizar cómo la probabilidad estimada converge al valor teórico.

```python
import numpy as np

# Definir los parámetros de la simulación
num_lanzamientos = 10_000 # Número de lanzamientos (simulaciones)

# Ejecutar todas las simulaciones a la vez
resultados = np.random.choice(["Cara", "Cruz"], size=num_lanzamientos)

# Calcular la probabilidad estimada de "Cara"
prob_cara = np.sum(resultados == "Cara") / num_lanzamientos

print(f"Probabilidad estimada de 'Cara': {prob_cara * 100:.2f}% (Teórica: 50%)")
# Calcular la robabilidad acumulada después de cada lanzamiento
frecuencia_acumulada = [np.sum(resultados[:n] == "Cara") / n for n in range(1, num_lanzamientos + 1)]

print(frecuencia_acumulada[-1])
```

En el código anterior, utilizamos la **Simulación Vectorizada** haciendo uso de “np.random.choice(...)”,  en lugar de un bucle, le pedimos a NumPy que "elija" 10,000 veces entre "Cara" y "Cruz" y nos devuelva todos. los resultados a la vez en un solo array. Esta es una forma de simulación extremadamente rápida y eficiente. Posteriormente, utilizamos “np.sum(resultados == "Cara") / num_lanzamientos”, la parte importante es “resultados == "Cara"”, ya que esta operación no compara un solo valor, sino que compara cada elemento del array resultados con la palabra "Cara". El resultado es un nuevo array de booleanos, con True donde había "Cara" y False donde había "Cruz".

#### **Ejemplo 2.**
En el siguiente ejemplo, abordamos una pregunta más interesante que requiere una simulación más elaborada: **"En promedio, ¿cuántos lanzamientos de moneda se necesitan para obtener 3 caras seguidas?"**. Para responder a esto con el método de Montecarlo, no basta con realizar el experimento una vez. Debemos hacerlo miles de veces y luego promediar los resultados. Para lograrlo, usaremos una estructura de **bucles anidados:**



1.   **Un bucle for externo** que se encargará de repetir el experimento completo un gran número de veces (por ejemplo, 10,000 simulaciones) para asegurar que nuestro resultado promedio sea fiable.
2.   **Un bucle while interno** que simulará los lanzamientos de moneda *dentro de un solo experimento* y continuará ejecutándose hasta que se alcance el objetivo de 3 caras seguidas.   



En el siguiente ejemplo, mostraremos como simular cuántos lanzamientos se necesitan para obtener **3 caras seguidas**, utilzamos un bucle for para iterar sobre un numero “n” de simulaciones, el guion bajo “_” lo utilizamos como variable de descarte (es una convención en Python para indicar que la variable del bucle no se usa dentro del bloque de código.), pero si ejecutamos las iteraciones. Posteriormente usamos el bucle “while” para iterar hasta que “caras_seguidas” es igual a tres.  

```python
import numpy as np

# Parámetros de la simulación
num_simulaciones = 1000

# Lista para guardar el número de lanzamientos de cada simulación
resultados = []

# Bucle Externo: Repite el experimento completo
for _ in range(num_simulaciones):
    lanzamientos = 0
    caras_seguidas = 0

    # Bucle Interno: Simula los lanzamientos hasta cumplir la condición
    while caras_seguidas < 3:
        lanzamientos += 1
        moneda = np.random.choice(['Cara', 'Cruz'])

        if moneda == 'Cara':
            caras_seguidas += 1
        else:
            caras_seguidas = 0

    resultados.append(lanzamientos)
  # Calcular promedio de lanzamientos
promedio = np.mean(resultados)


print(f"Después de {num_simulaciones} simulaciones:")
print(f"En promedio, se necesitan {promedio:.2f} lanzamientos para obtener 3 caras seguidas.")
```

En el código anterior, Creamos una lista vacía “resultados”, que se usa para almacenar el resultado (el número de lanzamientos) de cada una de las 10,000 simulaciones. El codigo “for _ in range(num_simulaciones)” inicia el bucle que se repetirá 10,000 veces. Utilizamos “_” como nombre de la variable temporal porque no necesitamos saber en qué número de simulación estamos (ej. la 1ª, 2ª o 500ª). Es una convención en Python para indicar que la variable del bucle se ignora intencionadamente.

Posteriormente utilizamos las variables “lanzamientos = 0” y “caras_seguidas = 0”. Estos contadores se reinician a cero al comienzo de cada nueva simulación del bucle “for”. Esto es fundamental para que cada experimento empiece de nuevo. El bucle “while” continuará ejecutándose mientras la racha de caras sea menor a 3. En cada vuelta del while, contamos un lanzamiento más (lanzamientos += 1), y si el resultado es "Cara" (if moneda == 'Cara':), incrementamos nuestro contador de racha (caras_seguidas). Si sale "Cruz", la racha se rompe. Esta es la parte clave, ya que el contador de “caras_seguidas” debe volver a cero, y el “while” continuará hasta que se consiga una nueva racha de 3 caras.  

Una vez que el bucle while termina (lo que solo ocurre cuando “caras_seguidas” finalmente llega a 3), la línea “resultados.append(lanzamientos)” guarda el número total de lanzamientos que se necesitaron para esa simulación específica.
