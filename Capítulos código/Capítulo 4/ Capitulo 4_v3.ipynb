### 4.1 Manipulación Avanzada de Arrays con NumPy

En el análisis económico de datos consiste en preparar y reestructurar los datos para que tengan la forma correcta que requieren los modelos o las visualizaciones. NumPy ofrece un conjunto de funciones potentes y flexibles para estas tareas. A continuación, exploramos las operaciones más comunes.

##### __Cambiar la forma de un array `reshape()`:__

Recibimos datos en un formato (como una larga lista 1D) y los convertimos a 2D o en otra forma compatible.

```python
import numpy as np


# Partimos de un array de 1D con 6 elementos
a = np.array([1, 2, 3, 4, 5, 6])

# Reestructuramos a una matriz de 2 filas y 3 columnas (2 * 3 = 6 elementos)
b = a.reshape((2, 3))


print("Array reestructurado:\n", b)
```

##### __Aplanar un array multidimensional `flatten()`:__

La operación inversa a “.reshape()” es “.flatten()”, que toma un array de cualquier dimensionalidad y lo convierte en un vector de una sola dimensión (1D).

```python
import numpy as np


# Usamos la matriz que creamos en el paso anterior
c = b.flatten()


print("Array aplanado:", c)
```

##### __Transponer un array transpose():__

La transposición de una matriz es una operación fundamental en álgebra lineal que consiste en intercambiar sus filas por sus columnas. En NumPy, esto es increíblemente simple usando el atributo "T".

```python
import numpy as np

matriz = np.array([[1, 2], [3, 4]])
print("Transpuesta:", matriz.T, sep="\n")
```

##### __Unir arrays `concatenate()`:__

Esta función nos permite combinar o "pegar" múltiples arrays en uno solo. La clave es el argumento axis, que define si la unión será vertical u horizontal.



1.   axis=0: une los arrays **verticalmente** (apila las filas). Los arrays deben tener el mismo número de columnas.
2.   axis=1: une los arrays **horizontalmente** (pega las columnas). Los arrays deben tener el mismo número de filas.



```python
import numpy as np

x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6]])

print("Concatenación vertical:", np.concatenate((x, y), axis=0), sep="\n")
```

### 4.2 Generación de Datos Aleatorios para Simulaciones Económicas

Una de las aplicaciones más poderosas de NumPy en la economía es la capacidad de generar datos aleatorios que sigan distribuciones de probabilidad específicas. Esto nos permite crear simulaciones de Montecarlo, probar la robustez de un modelo o generar datos sintéticos cuando no disponemos de datos reales.

El módulo “np.random” es nuestro centro de operaciones para esta tarea. A continuación, exploraremos tres de sus funciones más útiles.

##### __Distribución uniforme__:

La distribución uniforme se utiliza cuando asumimos que cualquier valor dentro de un rango [low, high) tiene exactamente la misma probabilidad de ocurrir. No hay un valor central o más probable.

1. low (primer parámetro): El valor más bajo
2. high (segundo parámetro): El valor más alto
3. size (tercer parámetro: El número de muestras a generar

```python
import numpy as np

# Simular tasas de interés entre 0.01 y 0.05
tasas = np.random.uniform(low=0.01, high=0.05, size=1000)

print("Primeras 5 tasas simuladas:", tasas[:5])
```

##### __Distribución normal (Gaussiana)__

Representa el clásico "patrón de campana", donde los valores se aglomeran alrededor de una media central y se vuelven cada vez más raros a medida que se alejan de ella. Es la distribución más importante en estadística.

Para simular este tipo de distribución en Python, NumPy hace uso de la función “np.random.normal(loc, scale, size)” la cual contiene tres parámetros que se muestran a continuación:  

1. loc (Primer parámetro): La Media (μ - mu) o Promedio de la Distribución
2. scale (Segundo parámetro): La Desviación Estándar (σ - sigma) de la Distribución.
3. size (Tercer parámetro): El Número de Muestras a Generar.

```python
import numpy as np

# Simular ingresos con distribución normal (media=30000, std=5000)
ingresos_simulados = np.random.normal(loc=30000, scale=5000, size=1000)

print("Primeros 5 ingresos simulados:", ingresos_simulados[:5])
```

##### __Enteros Aleatorios (Discreta Uniforme):__

Esta función genera números enteros aleatorios, donde cada entero dentro del rango especificado tiene la misma probabilidad de ser elegido.

Para simular este tipo de distribución en Python, NumPy hace uso de la función “np.random.randint(low, high, size)” la cual contiene tres parámetros que se muestran a continuación:
1. low (primer parámetro): El valor más bajo
2. high (segundo parámetro): El valor más alto
3. size (tercer parámetro: El número de muestras a generar

```python
import numpy as np

# Simular número de hijos por familia (entre 0 y 4)
hijos = np.random.randint(low = 0, high =5, size=10)

print("Hijos simulados por familia:", hijos)
```

### 4.3 Visualización de Datos con Matplotlib:

Las gráficas son herramientas fundamentales para comprender los datos, comunicar hallazgos y detectar patrones relevantes. Por ejemplo, un gráfico de líneas puede mostrar cómo cambia el PIB a lo largo del tiempo, mientras que un histograma puede revelar la distribución del ingreso en una población.

Python ofrece varias librerías para visualización, pero “Matplotlib” es una de las más utilizadas y poderosas. Es altamente flexible y permite crear desde gráficas sencillas hasta visualizaciones complejas. La estructura básica de “Matplotlib” se parece al proceso manual de hacer una gráfica en papel: se definen los datos a usar, el tipo de gráfica necesaria (líneas, barras, dispersión, etc.), y cómo etiquetar o personalizar los ejes y títulos.

Para utilizar la librería primero debemos instalarla, el comando para la instalación es “pip install matplotlib”,, posteriormente, importamos la libreria con “import matplotlib.pyplot as plt”; por convención, utilizamos el alias “plt”.

##### __Ejemplo 1: Gráfico de Línea: Crecimiento PIB__

Un gráfico de líneas es ideal para visualizar cómo una variable cambia a lo largo de una secuencia, típicamente el tiempo.

```python
import matplotlib.pyplot as plt

# 1. Preparar los Datos
años = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
pib_crecimiento = [2.3, 2.5, 2.8, 2.4, 1.8, -0.5, 3.1]

# 2. Crear el Gráfico
plt.plot(años, pib_crecimiento, marker='o', linestyle='-', color='b')



# 3. Personalizar y Etiquetar
plt.title("Crecimiento Anual del PIB (%)")
plt.xlabel("Año")
plt.ylabel("Tasa de Crecimiento")
plt.grid(True)

# 4. Mostrar el Gráfico
plt.show()
```

En el código anterior, primero importamos la librería “Matplotlib”, específicamente su módulo “pyplot”, que es el encargado de generar gráficos (se usa con el alias plt para escribir menos al llamarlo después). Posteriormente, creamos dos listas simples de Python. La primera, años, contiene los valores para nuestro eje horizontal (X). La segunda, pib_crecimiento, contiene los valores para nuestro eje vertical (Y). Es fundamental que ambas listas tengan el mismo número de elementos, ya que Matplotlib asocia cada punto de años con su correspondiente punto en pib_crecimiento (ej. 2015 se asocia con 2.3, 2016 con 2.5, y así sucesivamente).

##### __Ejemplo 2. Histograma — Distribución del Ingreso__

Un histograma nos permite ver la forma de la distribución de una única variable: dónde se concentran los valores, si es simétrica o sesgada, y si hay valores atípicos.

```python
import numpy as np

import matplotlib.pyplot as plt

# 1. Preparar los Datos (Simulados)
np.random.seed(42)
ingresos = np.random.normal(loc=30000, scale=7000, size=1000)

# 2. Crear el Gráfico
plt.hist(ingresos, bins=30, color='skyblue', edgecolor='black')

# 3. Personalizar y Etiquetar
plt.title("Distribución del Ingreso Simulado")
plt.xlabel("Ingreso Anual")
plt.ylabel("Frecuencia (Número de Personas)")


# 4. Mostrar el Gráfico
plt.show()
```

A diferencia del ejemplo anterior, aquí generamos 1,000 puntos de datos aleatorios que simulan una muestra de ingresos siguiendo una distribución normal. Esto nos da un conjunto de datos más denso para analizar.  

Posteriormente, utilizamos “plt.hist(...)”, esta es la función específica para crear histogramas. Su objetivo es agrupar los datos en rangos para visualizar su frecuencia y distribución. El primer argumento, ingresos, es el conjunto de datos que queremos analizar. El parámetro “bins=30” es clave en un histograma, ya que define en cuántos "contenedores" o "intervalos" se dividirán los datos a lo largo del eje “X”.

##### __Ejemplo 3: Gráfico de Dispersión — Consumo vs. Ingreso__

Este gráfico es fundamental para visualizar la relación o correlación entre dos variables numéricas, donde cada punto en el gráfico representa una observación


```python
import numpy as np
import matplotlib.pyplot as plt


# 1. Preparar los Datos (Simulados)
# El consumo es 70% del ingreso + ruido
ingreso = np.random.normal(30000, 5000, 100)
consumo = ingreso * 0.7 + np.random.normal(0, 2000, 100)


# 2. Crear el Gráfico
plt.scatter(ingreso, consumo, alpha=0.6, s=50) # s es el tamaño de los puntos


# 3. Personalizar y Etiquetar
plt.title("Relación entre Ingreso y Consumo")
plt.xlabel("Ingreso")
plt.ylabel("Consumo")
plt.grid(True)

# 4. Mostrar el Gráfico plt.show()
plt.show()
```

En el código anterior, se sigue la misma estructura de los dos ejemplos anteriores. Primero, generamos 100 observaciones para la variable ingreso a partir de una distribución normal. Luego, creamos la variable consumo de una manera que depende directamente del ingreso.

Posteriormente utilizamos “plt.scatter(...)” para crear el gráfico, donde el primer argumento (ingreso) corresponde al eje X, y el segundo (consumo) al eje Y. El parámetro “alpha=0.6” controla la transparencia de los puntos (1 es totalmente opaco, 0 es invisible). Usar un valor menor a 1 es muy útil cuando los puntos pueden superponerse, ya que nos permite identificar las áreas de mayor densidad. El parametro “s=50” controla el tamaño de los marcadores en el gráfico. Como en los casos anteriores, las funciones de personalización (plt.title, plt.xlabel, etc.) y la función final “plt.show()” nos permiten dar contexto a la visualización y mostrarla en pantalla

### Exportación de Gráficos:

Una vez que hemos creado una visualización, el paso final es guardarla como un archivo de imagen (PNG o PDF) para poder usarla en reportes, presentaciones o artículos. Matplotlib hace este proceso muy sencillo con la función “plt.savefig()”. Hay una regla de oro al guardar gráficos en Matplotlib que es fundamental: se debe llamar a “plt.savefig()” antes de llamar a “plt.show()”. La razón es que “plt.show()” muestra el gráfico en pantalla y, en muchos entornos, "limpia" la figura de la memoria después de hacerlo. Si se intenta guardar el gráfico después de mostrarlo, se podría terminar guardando un archivo en blanco.

```python
import matplotlib.pyplot as plt

# 1. Preparar los Datos
años = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
pib_crecimiento = [2.3, 2.5, 2.8, 2.4, 1.8, -0.5, 3.1]



# 2. Crear el Gráfico
plt.plot(años, pib_crecimiento, marker='o', linestyle='-', color='b')


# 3. Personalizar y Etiquetar
plt.title("Crecimiento Anual del PIB (%)")
plt.xlabel("Año")
plt.ylabel("Tasa de Crecimiento")
plt.grid(True)


# 4. GUARDAR el Gráfico (El paso nuevo y crucial)
# Se guarda ANTES de plt.show()
# El formato del archivo (.png, .pdf, .svg) se determina por la extensión
plt.savefig('grafico_crecimiento_pib.png', dpi=300, bbox_inches='tight')


#5. Mostrar el Gráfico en Pantalla
plt.show()
```

Al ejecutar el código anterior, no solo se mostrará el gráfico en el notebook, sino que también se guardará un nuevo archivo llamado “grafico_crecimiento_pib.png” en la misma carpeta donde se encuentra el cuaderno de Jupyter. La función “plt.savefig()” es muy flexible. Su primer argumento es siempre el nombre del archivo, y su extensión determina el formato, asi mismo, la funcion permite pasar algunos parametros para especificar la resolucion u otras opciones de la imagen, por ejemplo:  

- dpi=300: "Dots Per Inch" (Puntos por Pulgada). Controla la resolución de las imágenes rasterizadas (como PNG). Un valor de 300 es el estándar para una buena calidad de impresión.

- bbox_inches='tight': a menudo, Matplotlib guarda el gráfico con un borde blanco excesivo alrededor. Este argumento recorta automáticamente el espacio en blanco, dejando solo el gráfico.

- transparent=True: Si se añade, el fondo del gráfico será transparente en lugar de blanco, lo cual es perfecto si se quiere superponer sobre un fondo de color en una presentación.

### 4.4 Visualización de Datos con Seaborn:


Si bien Matplotlib nos da un control total sobre cada detalle de una figura, a menudo queremos crear visualizaciones estadísticas complejas (como comparaciones entre grupos o regresiones lineales visuales) de una manera más rápida y con una estética más pulida. Para esto, utilizamos Seaborn.

**1. Gráfico de Líneas: Estética Mejorada (lineplot)**

Usaremos los mismos datos del PIB. La sintaxis es muy similar a la de “plt.plot”, pero notaremos una apariencia visual diferente por defecto.

```python
import seaborn as sns
import matplotlib.pyplot as plt


# 1. Preparar los Datos
años = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
pib_crecimiento = [2.3, 2.5, 2.8, 2.4, 1.8, -0.5, 3.1]


# 2. Crear el Gráfico con Seaborn
sns.lineplot(x=años, y=pib_crecimiento, marker='o')



# 3. Personalizar y Mostrar (usando Matplotlib)
plt.title("Crecimiento Anual del PIB (con Seaborn)")
plt.xlabel("Año")
plt.ylabel("Tasa de Crecimiento (%)")
plt.grid(True)
plt.show()
```

En el código anterior, utilizamos “sns.lineplot”, el cual es el equivalente de Seaborn a “plt.plot”. Así mismo, hacemos uso de “Matplotlib” para extender la visualización del gráfico y agregar el título y etiquetas.  El resultado visual (estilo de línea, colores, fuentes) es más pulido que el solo utilizar Matplotlib.  

**2. Histograma con Curva de Densidad (histplot)**

Aquí veremos una de las grandes ventajas de Seaborn: la simplicidad para añadir capas estadísticas.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Preparar los Datos
np.random.seed(42)
ingresos = np.random.normal(loc=30000, scale=7000, size=1000)



# 2. Crear el Gráfico con Seaborn
# El argumento kde=True añade la curva de densidad automáticamente
sns.histplot(x=ingresos, bins=30, kde=True)


# 3. Personalizar y Etiquetar
plt.title("Distribución del Ingreso con Curva de Densidad (KDE)")
plt.xlabel("Ingreso Anual")
plt.ylabel("Frecuencia")


# 4. Mostrar el Gráfico
plt.show()
```

En el código anterior, utilizamos “sns.histplot”, que es la función de Seaborn para histogramas. La gran ventaja es el parámetro “kde=True”, que calcula y dibuja la curva de densidad sin necesidad de usar la librería SciPy, haciendo el proceso mucho más directo y el código más limpio.

**3. Gráfico de Dispersión con Línea de Regresión (regplot)**

El siguiente ejemplo, muestra el verdadero poder estadístico de Seaborn. Con una sola función, creamos un gráfico de dispersión y le superponemos un modelo de regresión lineal

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Preparar los Datos
np.random.seed(1) # Usamos una semilla para que el ejemplo sea reproducible
ingreso = np.random.normal(30000, 5000, 100)
consumo = ingreso * 0.7 + np.random.normal(0, 2000, 100)



# 2. Crear el Gráfico de Regresión con Seaborn
sns.regplot(x=ingreso, y=consumo, line_kws={"color":"red"},
scatter_kws={'alpha':0.5})



# 3. Personalizar y Etiquetar
plt.title("Relación entre Ingreso y Consumo con Regresión")
plt.xlabel("Ingreso")
plt.ylabel("Consumo")
plt.grid(True)


# 4. Mostrar el Gráfico
plt.show()
```

En el código anterior utilizamos “sns.regplot” que dibuja la gráfica de dispersión de los puntos (ingreso vs. consumo), calcula automáticamente la línea de regresión lineal que mejor se ajusta a esos puntos (la hemos puesto roja con “line_kws”); muestra el intervalo de confianza del 95% para la estimación de la regresión como un área sombreada. Lograr esto con Matplotlib requeriría cálculos estadísticos adicionales (probablemente con scipy o statsmodels) y varias líneas de código de graficación. Seaborn lo abstrae todo en una sola y potente función.
