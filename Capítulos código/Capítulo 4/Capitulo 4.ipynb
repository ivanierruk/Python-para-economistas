### 4.1 Manipulación Avanzada de Arrays con NumPy

El análisis económico con datos requiere transformar arrays de manera flexible, exploramos funciones útiles para esas tareas.

##### __Cambiar la forma de un array `reshape()`__

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape((2, 3))

print("Array reestructurado:", b, sep="\n")
```

##### __Aplanar un array multidimensional `flatten()`__

```python
import numpy as np

c = b.flatten()
print("Array aplanado:", c)
```

##### __Transponer un array transpose()__

```python
import numpy as np

matriz = np.array([[1, 2], [3, 4]])
print("Transpuesta:", matriz.T, sep="\n")
```

##### __Unir arrays `concatenate()`__

```python
import numpy as np

x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6]])

print("Concatenación vertical:", np.concatenate((x, y), axis=0), sep="\n")
```

### 4.2 Generación de Datos Aleatorios para Simulaciones Económicas

 NumPy permite generar datos aleatorios útiles para simular ingresos, precios o shocks económicos:

##### __Distribución uniforme__

```python
# Simular tasas de interés entre 0.01 y 0.05
tasas = np.random.uniform(low=0.01, high=0.05, size=1000)

print("Primeras 5 tasas simuladas:", tasas[:5])
```

##### __Distribución normal__

1. loc (Primer parámetro): La Media (μ - mu) o Promedio de la Distribución
2. scale (Segundo parámetro): La Desviación Estándar (σ - sigma) de la Distribución.
3. size (Tercer parámetro): El Número de Muestras a Generar.

```python
import numpy as np

# Simular ingresos con distribución normal (media=30000, std=5000)
ingresos_simulados = np.random.normal(loc=30000, scale=5000, size=1000)

print("Primeros 5 ingresos simulados:", ingresos_simulados[:5])
```

##### __Distribuciones discretas: enteros aleatorios__

```python
import numpy as np

# Simular número de hijos por familia (entre 0 y 4)
hijos = np.random.randint(0, 5, size=10)

print("Hijos simulados por familia:", hijos)
```

### 4.3 Visualización de Datos con Matplotlib

##### __Ejemplo 1: Gráfico de Línea — Crecimiento del PIB__

```python
import matplotlib.pyplot as plt

años = [2015, 2016, 2017, 2018, 2019, 2020]
pib = [2.3, 2.5, 2.8, 2.4, 1.8, -0.5]

plt.plot(años, pib, marker='o')
plt.title("Crecimiento del PIB")
plt.xlabel("Año")
plt.ylabel("Tasa de crecimiento (%)")
plt.grid(True)
plt.show()
```

##### __Ejemplo 2. Histograma — Distribución del Ingreso__

```python
import numpy as np
import matplotlib.pyplot as plt

ingresos = np.random.normal(loc=30000, scale=5000, size=1000)

plt.hist(ingresos, bins=30, color='skyblue', edgecolor='black')
plt.title("Distribución del Ingreso Simulado")
plt.xlabel("Ingreso")
plt.ylabel("Frecuencia")
plt.show()
```

##### __Ejemplo 3: Gráfico de Dispersión — Consumo vs. Ingreso__



```python
import numpy as np
import matplotlib.pyplot as plt

ingreso = np.random.normal(loc=30000, scale=5000, size=100)
consumo = ingreso * 0.8 + np.random.normal(loc=0, scale=1000, size=100)

plt.figure(figsize=(8, 6))
plt.scatter(ingreso, consumo, alpha=0.7, s=15)
plt.title("Relación entre Ingreso y Consumo (Histograma Bidimensional)")
plt.xlabel("Ingreso")
plt.ylabel("Consumo")
plt.grid(True)
plt.show()
```
